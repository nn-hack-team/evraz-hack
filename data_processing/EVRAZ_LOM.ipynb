```python
lom_test = pd.read_csv('/content/lom_test.csv')
```

```python
def preparation(lom_train):
  lom_train = pd.get_dummies(lom_train, columns=['NML'])
  lom_train = pd.get_dummies(lom_train, columns=['VDL'])
  lom_train = lom_train.groupby('NPLV').agg({'VES': 'sum',
                               'NML_25КШ': 'sum',
                               'NML_К   ': 'sum',
                               'NML_КП  ': 'sum',
                               'NML_КШС8': 'sum',
                               'NML_ЛЧ  ': 'sum',
                               
                               'NML_О   ': 'sum',
                               'NML_ОК  ': 'sum',
                               'NML_СК  ': 'sum', 'NML_У2КШ' : 'sum',
                             
                             'VDL_4':'sum',
                             'VDL_8':'sum', 
                             'VDL_13':'sum', 
                             'VDL_20':'sum', 
                             'VDL_23':'sum', 
                             'VDL_48':'sum',
                             'VDL_49':'sum',
                              'VDL_61':'sum',
                               'VDL_63':'sum' })
  lom_train['VES'] = scaler.fit_transform(np.array(lom_train['VES']).reshape(-1,1))
  return lom_train
```

```python
lom_test = preparation(lom_test)
```

```python
lom_test = lom_test.reset_index()
```

```python
lom_test.to_csv('lom_test_csv.csv', index=False)
```
